```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
# ========== 1. Load the dataset ==========
data=pd.read_csv("coffeeshop_dataset.csv",parse_dates=['SaleDate'])
```

# Display basic information

```python
data.describe()
```

```python
print("\nNull Values:\n",data.isnull().sum())
```

```python
data['Profit'].value_counts().plot(kind='pie',figsize=(14,6))
```

#Data Visualization

```python
#The mean of customer Age
data['CustomerAge'].mean()
```

```python
#The density of customer Age
data['CustomerAge'].plot(kind='kde',figsize=(14,6))
```

```python
#Profit by SaleDate
plt.figure(figsize=(14,6))
sns.lineplot(data=data,x="SaleDate",y="Profit",errorbar=None)
plt.title("Total Profit by SaleDate")
plt.xlabel("SaleDate")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.show()
```

```python
#The mean of unit sold
data['UnitsSold'].mean()
```

```python
#Histogram for the unitsold
data['UnitsSold'].plot(kind='hist',figsize=(10,6),bins=10)
```

```python
#Day of the week 
data['DayOfWeek'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=data,x="DayOfWeek",y="Profit",ci=None,estimator=sum)
plt.title("Total Profit by day of week")
plt.show()
```

```python
#Profit Frequency
data['Profit'].values
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=data,x="Profit",bins=30,kde=True)
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()
```

```python
#Sales per day
data['SaleDate'].value_counts()
```

```python
#Category
DPC=data['Category'].value_counts()
DPC
```

```python
DPC.plot(kind='pie',figsize=(10,6))
```

```python
#Profit by category
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x="Category", y="Profit")
plt.title("Profit by Item Category")
plt.xticks(rotation=45)
plt.show()
```

```python
#scatter plot for the category vs the profit
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="Price", y="Profit", hue="Category")
plt.title("Profit vs. Price by Category")
plt.show()
```

```python
#Hexagonal Bin Plot for the prive vs profit
plt.figure(figsize=(10,6))
plt.hexbin(data['Price'],data['Profit'],gridsize=30,cmap="Blues")
plt.colorbar(label="Count")
plt.title("Hexbin Plot of Profit vs Price")
plt.xlabel('Price')
plt.ylabel('Profit')
plt.show()
```

```python
#Code for every product sold
data.loc[:,'ItemName'].unique()
```

```python
#Best sellers
data['ItemName'].value_counts().head(1)
```

```python
#Profit grouped by category
CP=data.groupby("Category")["Profit"].sum()
plt.figure(figsize=(8,8))
CP.plot.pie(autopct="%1.1f%%",startangle=90,cmap="Pastel1")
plt.title("Profit by Category")
plt.ylabel("")
plt.show()
```
